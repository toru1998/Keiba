```python
def summarize_sqlite_db(db_path: str):
    """
    SQLite DB ã®ã‚µãƒãƒªãƒ¼æƒ…å ±ã‚’è¡¨ç¤ºã™ã‚‹é–¢æ•°

    Parameters
    ----------
    db_path : str
        SQLite3 ã® DB ãƒ•ã‚¡ã‚¤ãƒ«ã®ãƒ‘ã‚¹
    """
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    print(f"ğŸ“Š Database: {db_path}")

    # ãƒ†ãƒ¼ãƒ–ãƒ«ä¸€è¦§å–å¾—
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = [row[0] for row in cursor.fetchall()]
    print(f"ãƒ†ãƒ¼ãƒ–ãƒ«æ•°: {len(tables)}")
    print(f"ãƒ†ãƒ¼ãƒ–ãƒ«ä¸€è¦§: {tables}")
    print("="*60)

    for table in tables:
        print(f"ğŸ“Œ Table: {table}")

        # ã‚«ãƒ©ãƒ æƒ…å ±
        cursor.execute(f"PRAGMA table_info({table});")
        columns = cursor.fetchall()
        print("ã‚«ãƒ©ãƒ :")
        for col in columns:
            cid, name, col_type, notnull, dflt_value, pk = col
            print(f"  - {name} ({col_type}), NOT NULL={bool(notnull)}, PK={bool(pk)}")

        # ãƒ¬ã‚³ãƒ¼ãƒ‰ä»¶æ•°
        cursor.execute(f"SELECT COUNT(*) FROM {table};")
        row_count = cursor.fetchone()[0]
        print(f"ãƒ¬ã‚³ãƒ¼ãƒ‰ä»¶æ•°: {row_count}")

        print("-"*60)

    conn.close()


# ä¾‹: ä½¿ã„æ–¹
if __name__ == "__main__":
    summarize_sqlite_db("../data/db/keiba.db")
```
